## What is the age distribution of the passengers of the Titanic?

````python
heights = titanic_df["Age"] #turn Age into list

plt.figure(figsize=(10, 6), facecolor="lightblue") #set figure size + outer background color
plt.hist(heights, bins=8, edgecolor='black', color="red") #draw histogram
plt.title("Age Distribution of The Titanic's Passengers") # write title
plt.xlabel('Age') # write x axis label
plt.ylabel('Number of Passager') #write y axis label
plt.gca().set_facecolor('lightgray') #set inner background to lightgray
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s)
````

## Compareed to third and second class how many were riding first class?

````python
pclass = titanic_df['Pclass'].value_counts() #get amount of each class
labels = ["3rd", "1st", "2nd"] #set labels

plt.figure(figsize=(10,10), facecolor="lightblue") #set figure size + outer background color

plt.pie(pclass, labels = labels) #plot pie chart

plt.title("Passengers Classes") #write title
````

### Written Analysis:
The graphs show that most passengers were in third class, while the remaining were second

## Did males pay more for the fares than females?

````python
male = titanic_df[titanic_df["Sex"] == "male"] #filter for males
female = titanic_df[titanic_df["Sex"] == "female"] #filter for females

m_fares = male["Fare"] #make list of male fares
f_fares = female["Fare"] #make list of female fare

m_a_fare = m_fares.mean() #find mean of male fares
f_a_fare = f_fares.mean() #find mean of female fares

a_fare = [m_a_fare, f_a_fare] #turn data into list for easier input
sex = ["Male", "Female"] #turn label into list

plt.figure(figsize=(10, 6), facecolor="lightblue") #set figure size + outer background color
plt.bar(sex, a_fare, color="purple") #draw bar graph
plt.title('Average Fair Difference between Sexes') #write title
plt.xlabel('Sex') #write x label
plt.ylabel('Average Fair') #write y label

plt.gca().set_facecolor('lightgray') #set inner background to lightgray
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s) active figure objects, and opens >=1 interactive windows that display figure(s)
````

### Written Analysis:
No, males on average did not not pay more the females, in fact on average females paid more for their fare than males.

## Did having siblings or spouses increase survival rate?

````python
survival = titanic_df.groupby(["SibSp", "Survived"])[["PassengerId"]].count().unstack()

plt.figure(facecolor="lightblue") #set outer background to lightblue
#figure() must go before everything
survival.plot.bar(figsize=(16, 10), color=["red", "green"])
#everything elses goes after plot
plt.title('Average Fair Difference between Sexes') #write title
plt.xlabel('Number of Siblings or Spouses') #write x label
plt.ylabel('Survival Rate') #write y label
plt.gca().set_facecolor('lightgray') #set inner background to lightgray
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s) active figure objects, and opens >=1 interactive windows that display figure(s)
````

### Written Analysis:
The graphs show that people with 1 or 2 siblings or spouses on board with then are more likey to survive. The graphs also shows that the survival rate start to decrease after the number of siblings or spouses goes beyond 2. The graphs also shows that most people had zero siblings or spouses on board with them and those with 0 siblings or spouses on board had a round a 50% change of surviving.

## Did passengers' class and sex affect survival rate?

````python
survival = titanic_df.groupby([ "Pclass", "Sex", "Survived"])[["PassengerId"]].count().unstack() #create unstacked groupby list

plt.figure(facecolor="blue") #set figure color to light blue
survival.plot.bar(figsize=(18, 10), color=["red", "green"]) #create graph with figsize + color
plt.title('Average Fair Difference between Sexes') #write title
plt.xlabel('Passager Class') #write x label
plt.ylabel('Survival Rate') #write y label

plt.gca().set_facecolor('lightgray') #set inner background to lightgray
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s) active figure objects, and opens >=1 interactive windows that display figure(s)
````

### Written Analysis:
The graph shows that the higher the passenger's the more likely they are able to survive and that females were more likey to survive than males.
