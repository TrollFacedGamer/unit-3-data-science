## What is the age distribution of the passengers of the Titanic?

````python
heights = titanic_df["Age"] #turn Age into list

plt.figure(figsize=(10, 6)) #set figure size
plt.hist(heights, bins=8, edgecolor='black') #draw histogram
plt.title("Age Distribution of The Titanic's Passengers") # write title
plt.xlabel('Age') # write x axis label
plt.ylabel('Number of Passager') #write y axis label
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s)
````

## What percentage of passengers were riding first class?

````python
pclass = titanic_df['Pclass'].value_counts() #get amount of each class
labels = ["3rd", "1st", "2nd"] #set labels

plt.pie(pclass, labels = labels) #plot pie chart

plt.title("Passengers Classes") #write title

````

## Did males pay more for the fares than females?

````python
male = titanic_df[titanic_df["Sex"] == "male"]
female = titanic_df[titanic_df["Sex"] == "female"]

m_fares = male["Fare"] #list of data samples
f_fares = female["Fare"]

m_a_fare = m_fares.mean()
f_a_fare = f_fares.mean()

a_fare = [m_a_fare, f_a_fare]
sex = ["Male", "Female"]

plt.figure(figsize=(10, 6)) # set figure size
plt.bar(sex, a_fare) #draw bar graph
plt.title('Time Spend After School') #write title
plt.xlabel('Sex') #write x label
plt.ylabel('Average Fair') #write y label
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s) active figure objects, and opens >=1 interactive windows that display figure(s)
````
