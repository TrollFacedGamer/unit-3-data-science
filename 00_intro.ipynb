Now let's change the plot area (inner) background

```python
#set figure size to 10 wide 6 long
plt.figure(figsize=(10, 6))
ax = plt.gca()  # get current axes
ax.set_facecolor('lightblue') #set inner background to lightblue
plt.plot([1, 2, 3, 4], [1, 4, 2, 3]) #draw line graph
plt.title('Yellow Plot Area Background') #add title
```

#### 2.0 - Line Plot
Let's start with a simple line plot.

```python
# list for x values
month = [1, 2, 3, 4, 5, 6]
#list for y values
sales = [2, 4, 5, 8, 9, 12]

#set figure size
plt.figure(figsize=(20, 10))
#draw line graph
plt.plot(month, sales)
#write title
plt.title('Monthly Sales')
plt.xlabel('Month') #write x axis label
plt.ylabel('Sales (thousands)') #write y axis label
plt.grid(True) # draw grid
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s)
```

#### 2.1 - Scatter Plot
Let's look at some data about students' study time and test scores:

```python
study_hours = [1, 2, 2, 3, 4, 4, 5, 5, 6] #list for x values
test_scores = [65, 70, 75, 80, 85, 85, 90, 95, 100] # list for y values

plt.figure(figsize=(10, 6)) #set figure size
plt.scatter(study_hours, test_scores) #draw scatter plot
plt.title('Study Time vs. Test Scores') #write title
plt.xlabel('Hours Studied') #write x axis label
plt.ylabel('Test Score') # write y axis label
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s)
```

#### 2.3 - Histogram
Let's look at the distribution of heights in a class:

```python
heights = [65, 68, 67, 67, 70, 65, 68, 69, 66, 71, 
           67, 68, 68, 69, 70, 67, 66, 67, 68, 69] #list of data samples

plt.figure(figsize=(10, 6)) #set figure size
plt.hist(heights, bins=7, edgecolor='black') #draw histogram
plt.title('Student Heights') # write title
plt.xlabel('Height (inches)') # write x axis label
plt.ylabel('Number of Students') #write y axis label
plt.show() #starts loop, find currently active figure objects, and opens >=1 interactive windows that display figure(s)
```
